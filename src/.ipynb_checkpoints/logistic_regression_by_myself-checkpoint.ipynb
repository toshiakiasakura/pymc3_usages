```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set(context="paper" , style ="whitegrid",rc={"figure.facecolor":"white"})

import statsmodels.api as sm
import pymc3 as pm
import theano
import theano.tensor as tt

import warnings
warnings.filterwarnings('ignore')
```

# Logistic Regression with PyMC3  

```python
# data preparation 
m = len(exog)
x = df[exog].values
y = df[endog].values
x_shared = theano.shared(x)

# sampling parameters 
nIter = 500
nBurn = 500
chains = 4
cores = 2
randomSeed = 1123890

with pm.Model() as model:
    coeff = pm.Normal('beta', mu=0, sigma=100,shape=m)
    line = tt.sum(coeff * x_shared, axis=1)
    logistic = pm.math.sigmoid( line )
    pm.Bernoulli('obs', p=logistic, observed=y)

    # fit the model
    trace = pm.sample(draws=nIter,
                      tune=nBurn, 
                      chains=chains,
                      cores=cores,
                      random_seed=randomSeed)
```

##  check the result

```python
alpha = 0.95 # for credible interval
```

```python
# NOTE: just pm.traceplot behaves oddly. so that I use this implementation.
for i in range(m):
    pm.traceplot(trace,var_names="beta",coords={"beta_dim_0":[i]})
```

```python
pm.summary(trace,credible_interval=alpha).round(3)
```

```python
pm.plot_forest(trace,
               kind='ridgeplot',
               var_names=['beta'],
               combined=False,
               ridgeplot_overlap=4,
               ridgeplot_alpha=.25,
               colors='black',
               figsize=(9, 7))
```

```python
for i in range(2):
    pm.plot_posterior(trace, var_names=['beta'],coords ={"beta_dim_0":[i]},credible_interval=alpha)
```

```python
pm.forestplot(trace,credible_interval=alpha)
```

```python
pm.energyplot(trace)
```

```python
pm.pairplot(trace, var_names=['beta'], divergences=True);
```

```python
# obtain original samples. 
beta = trace["beta"]
np.mean(beta,axis=0)
```

#  Logistic Regression with PyMC3 using custom function


```python
EPS = np.finfo(float).eps
def logiFunc(x):
    v = 1./(1. + tt.exp(-x) )
    v = tt.clip(v,EPS,1-EPS)
    return(v)
 
def pred(X,w):
    linear = tt.sum(w*X,axis=1)
    y =  logiFunc(linear)  
    return(y)

def logLikelihood(t,X,w):
    y = pred(X,w) 
    ll = t * tt.log(y) + (1-t)*tt.log(1-y) 
    ll = tt.sum(ll)
    return(ll)
```

```python
# data preparation 
df = sm.datasets.anes96.load_pandas().data
exog = ["TVnews","selfLR","PID","educ"]
endog = "vote"

# set data 
m = len(exog)
x = df[exog].values
y = df[endog].values
x_shared = theano.shared(x)
# sampling parameters 

nIter = 500
with pm.Model() as model2:
    coeff = pm.Normal('beta', mu=0, sigma=100,shape=m)
    pm.DensityDist('likelihood',logLikelihood , observed=dict(t=y,X=x,w=coeff))

    # fit the model
    trace2 = pm.sample(draws=nIter)
```

```python
pm.summary(trace2,credible_interval=alpha).round(3)
```

#  Record reuslts of PyMC3 
data are saved using pickle. 

```python
import pickle
import datetime 
```

```python
# data preparation 
df = sm.datasets.anes96.load_pandas().data
exog = ["TVnews","selfLR","PID","educ"]
endog = "vote"

# set data 
m = len(exog)
x = df[exog].values
y = df[endog].values
x_shared = theano.shared(x)

# sampling parameters 
nIter = 500

with pm.Model() as model3:
    coeff = pm.Normal('beta', mu=0, sigma=100,shape=m)
    line = tt.sum(coeff * x_shared, axis=1)
    logistic = pm.math.sigmoid( line )
    pm.Bernoulli('obs', p=logistic, observed=y)

    # fit the model
    trace3 = pm.sample(draws=nIter)
# save result using pickle.
time = datetime.datetime.now().strftime("%Y%m%d%H%M")
savePath = "./model_%s.db" % time 

with open(savePath, 'wb') as buff:
    pickle.dump({'model': model3, 'trace': trace3, 'x_shared': x_shared}, buff)
```

```python
# reuse the result. 
with open(savePath, 'rb') as buff:
    data = pickle.load(buff)
model4 = data['model']
trace4 = data['trace']
x_shared = data['x_shared']
x_shared.set_value(np.asarray(x,theano.config.floatX))
```

```python
for i in range(m):
    pm.traceplot(trace4,var_names="beta",coords={"beta_dim_0":[i]})
```

#  Logistic regressoin with  class
please check also this site, https://docs.pymc.io/notebooks/blackbox_external_likelihood.html 

```python
EPS = np.finfo(float).eps
def logiFunc(x):
    v = 1./(1. + tt.exp(-x) )
    v = tt.clip(v,EPS,1-EPS)
    return(v)

class LogLike(tt.Op):
    itypes = [tt.dvector] # expects a vector of parameter values when called
    otypes = [tt.dscalar] # outputs a single scalar value (the log likelihood)

    def __init__(self, exog,endog):
        self.Phi  = tt.as_tensor_variable(exog)
        self.t = tt.as_tensor_variable(endog)
        
    def pred(self):
        linear = tt.sum(self.w * self.Phi,axis=1)
        self.y =  logiFunc(linear)   # N*1 vector 
        
    def logp(self,theta):
        self.w = tt.flatten(theta)
        self.pred()
        t = self.t
        y = self.y
        ll = t * tt.log(y) + (1-t)*tt.log(1-y) 
        ll = tt.sum(ll)
        return(ll)

    def perform(self, node, inputs, outputs):
        theta   = inputs # this will contain my variables
        logl = self.logp(theta)
        outputs[0][0] = np.array(logl) # output the log-likelihood
```

```python
# data preparation 
df = sm.datasets.anes96.load_pandas().data
exog = ["TVnews","selfLR","PID","educ"]
endog = "vote"

# set data 
x = df[exog].values
y = df[endog].values

# create our Op
nIter = 500

# use PyMC3 to sampler from log-likelihood
with pm.Model() as model4:
    # for illustration of multiple parameters,
    # I divide parameters into two parts. 
    beta1 = pm.Normal('beta1', mu=0, sigma=100,shape=2)
    beta2 = pm.Normal('beta2', mu=0, sigma=100,shape=2)    
    theta = tt.as_tensor_variable([beta1,beta2])
    logp = LogLike(x,y).logp(theta) # two parametrs are not accepted. 
    pm.Potential("likelihood",logp)

    trace4 = pm.sample(draws=nIter)
```

```python
pm.model_to_graphviz(model4)
```

```python
alpha = 0.95
pm.summary(trace4,credible_interval=alpha).round(3)
```
